```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

candy = pd.read_csv('../input/fivethirtyeight-candy-power-ranking-dataset/candy-data.csv')

label = candy[['competitorname']]
X = candy.drop(['competitorname', 'winpercent'], axis=1).to_numpy()
y = candy[['winpercent']].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.59, random_state=0)
```

```python
model = RandomForestRegressor()
model.n_estimators = 100
model.max_depth = 6
model.random_state = 1
model.fit(X_train, y_train)

print(f'Train MAE : {mean_absolute_error(model.predict(X_train), y_train)}')
print(f'Test MAE : {mean_absolute_error(model.predict(X_test), y_test)}')
```

```python
for n, i in zip(candy.drop(['competitorname', 'winpercent'], axis=1).columns, model.feature_importances_):
    print(n, ':', round(i, 2))
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error

tic = pd.read_csv('../input/tictactoe-endgame-dataset-uci/tic-tac-toe-endgame.csv')

X = tic.drop(['V10'], axis=1).replace({'x':1.0,'o':0.0, 'b':0.0}).to_numpy()
y = tic[['V10']].apply(lambda x : np.where(x=='positive',1.0, 0.0)).to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, random_state=0)
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

## Model metrics

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score

y_pred = model.predict(X_test)
confusion_matrix(y_pred, y_test)
```

```python
print((37+59)/(37+59)); print(accuracy_score(y_pred, y_test))
```

```python
print((59/59)); print(precision_score(y_pred, y_test))
```

```python
print(59/59); print(recall_score(y_pred, y_test))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, make_scorer


candy = pd.read_csv('../input/fivethirtyeight-candy-power-ranking-dataset/candy-data.csv')

label = candy[['competitorname']]
X = candy.drop(['competitorname', 'winpercent'], axis=1).to_numpy()
y = candy[['winpercent']].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.59, random_state=0)
```

```python
rfc = RandomForestRegressor()
mse = make_scorer(mean_absolute_error)

cv = cross_val_score(rfc, X_train, y_train, cv=10, scoring=mse)
```
