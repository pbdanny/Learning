```python
import pandas as pd

df = pd.DataFrame({
    'date' : ['05/03', '06/03', '07/03', '08/03'],
    'AA' : [1, 4, 7, 5],
    'BB' : [2, 5, 8, 7],
    'CC' : [3, 6, 9, 1]
}).set_index('date')
```

```python
df
```

```python
df.unstack().reset_index().rename(columns={'level_0':'gr', '0':'val'})
```

```python
df.reset_index().melt(id_vars='date', value_vars=['AA', 'BB', 'CC'])
```
