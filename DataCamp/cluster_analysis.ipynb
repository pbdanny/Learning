```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.DataFrame({'x':[9.0, 6, 2, 3, 1, 7, 1, 6, 1, 7, 23, 26, 25, 23, 21, 23, 23, 20, 30, 23],
                   'y':[8.0, 4, 10, 6, 0, 4, 10, 10, 6, 1, 29, 25, 30, 29, 29, 30, 25, 27, 26, 30]})
```

```python
plt.scatter(df['x'], df['y'])
```

```python
from scipy.cluster.hierarchy import linkage, fcluster
Z = linkage(df, 'ward')
n_clust = 2
cluster_labels = fcluster(Z, n_clust, criterion='maxclust')
plt.scatter(x=df['x'], y=df['y'], c=cluster_labels)
plt.show()
```

```python
from scipy.cluster.vq import kmeans, vq
centroids,_ = kmeans(df, 2)
df['cluster_labels'], _ = vq(df, centroids)
plt.scatter(x=df['x'], y=df['y'], c=df['cluster_labels'])
plt.show()
```

```python
from scipy.cluster.vq import whiten
goals_for = [4,3,2,3,1,1,2,0,1,4]
scaled_data = whiten(goals_for)
print(scaled_data)
plt.plot(goals_for, label='original')
plt.plot(scaled_data, label='scaled')
plt.legend()
plt.show()
```
