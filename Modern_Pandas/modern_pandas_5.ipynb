```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_rows = 30
sns.set(style='ticks', context='talk')
```

```python
fp = './data/nba.csv'

if not os.path.exists(fp):
    tables = pd.read_html("http://www.basketball-reference.com/leagues/NBA_2016_games.html")
    games = tables[0]
    games.to_csv(fp)
else:
    games = pd.read_csv(fp)
```

```python
column_names = {'Date': 'date', 'Start (ET)': 'start',
                'Unamed: 6': 'box', 'Visitor/Neutral': 'away_team', 
                'PTS': 'away_points', 'Home/Neutral': 'home_team',
                'PTS.1': 'home_points', 'Unamed: 7': 'n_ot'}
```

```python
games = (
    games
    .rename(columns=column_names)
    .dropna(thresh=4)
    [['date', 'away_team', 'away_points', 'home_team', 'home_points']]
    .assign(date = lambda x : pd.to_datetime(x['date'], format = '%a, %b %d, %Y'))
    .set_index('date', append=True)
    .rename_axis(['game_id', 'date'])
    .sort_index()
    )
```

```python
games.head()
```

```python
rest = (pd
 .melt(games.reset_index(), 
       id_vars=['game_id', 'date'], 
       value_vars = ['away_team', 'home_team'], 
       value_name='team')
 .sort_values('date')
 .assign(lag_date = lambda x : x.groupby('team')['date'].transform(pd.Series.shift, 1)) # lag date with shift, if index not date, use index sequence
 .assign(lag_days = lambda x : (x['date'] - x['lag_date']).dt.days - 1) # convert dt to float
 .assign(lag_days_2 = lambda x : (x.groupby(['team'])['date'].transform(pd.Series.diff)).dt.days - 1) # .diff, use index sequence
)
```

```python
rest = rest.assign(rest_days = lambda x : x.sort_values('date').groupby(['team'])['date'].diff().dt.days - 1)
```

```python
rest.dropna().head()
```

## Mini-Project : Probability to win for Home team

```python
df['home_win'] = df['home_points'] > df['away_points']
```

```python
df.head()
```

```python
wins = \
(
pd.melt(df.reset_index(),
       id_vars=['game_id', 'date', 'home_win'],
       value_vars = ['home_team', 'away_team'],
       var_name='is_home', 
       value_name='team')
    .assign(win = lambda x : x.home_win == (x.is_home == 'home_team'))
    .groupby(['team', 'is_home'])
    .win
    .agg(['sum', 'count', 'mean'])
    .rename(columns = dict(sum= 'n_wins', count= 'n_games', mean= 'win_pct'))
)
```

```python
g = sns.FacetGrid(wins.reset_index(), hue='team', aspect=.5, palette=['k'], height = 7)
g.map(sns.pointplot, 'is_home', 'win_pct', order =['away_team', 'home_team']).set(ylim=(0, 1));
```

```python
g = sns.FacetGrid(wins.reset_index(), col='team', hue='team', col_wrap=5, height = 2)
g.map(sns.pointplot, 'is_home', 'win_pct', order = ['away_team', 'home_team'])
```

```python
win_percent = (
    # Use sum(games) / sum(games) instead of mean
    # since I don't know if teams play the same
    # number of games at home as away
    wins.groupby(level='team', as_index=True)
        .apply(lambda x: x.n_wins.sum() / x.n_games.sum())
)
win_percent.head()
```

```python
win_percent.sort_values().plot.barh(figsize=(6, 12), width=.85, color='k')
plt.tight_layout()
sns.despine()
plt.xlabel("Win Percent")
```

```python
plt.figure(figsize=(8, 5))
(wins.win_pct
    .unstack()
    .assign(**{'Home Win % - Away %': lambda x: x.home_team - x.away_team,
               'Overall %': lambda x: (x.home_team + x.away_team) / 2})
     .pipe((sns.regplot, 'data'), x='Overall %', y='Home Win % - Away %')
)
sns.despine()
plt.tight_layout()
```

```python
df = df.assign(away_strength=df['away_team'].map(win_percent),
               home_strength=df['home_team'].map(win_percent),
               point_diff=df['home_points'] - df['away_points'],
               rest_diff=df['home_rest'] - df['away_rest'])
df.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

df['home_win'] = df.home_win.astype(int)  # for statsmodels
```

```python
# convert df to series
Y = df['home_win']
X = sm.add_constant(df[['home_strength', 'away_strength', 'home_rest', 'away_rest']])

model = sm.Logit(Y, X)
result = model.fit()
result.summary()
```

```python
mod = smf.logit('home_win ~ home_strength + away_strength + home_rest + away_rest', df)
res = mod.fit()
res.summary()
```
