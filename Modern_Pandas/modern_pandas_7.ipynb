### Time Series Modeling

```python
store = 'data/flights.h5'
df = (
    pd
    .read_hdf(store)
    [['fl_date', 'origin', 'crs_dep_time', 'dep_time', 'crs_arr_time', 'arr_time']]
)
```

```python
daily = df.fl_date.value_counts().sort_index()
```

```python
daily = df.fl_date.value_counts().sort_index()
y = daily.resample('2d').mean()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.tsa.api as smt
```

```python
X = pd.concat([y.shift(i) for i in range(6)], axis = 1, keys = ['y'] + ['L{}'.format(i) for i in range(1, 6)])
```

```python
formula = 'y ~ trend + L1 + L2 + L3 + L4 + L5'
```

```python
model = smf.ols(formula=formula, data = X.assign(trend = np.arange(len(X)))).fit()
```

```python
model.summary()
```

```python
sns.heatmap(X.corr(), cmap = 'PuBu')
plt.show()
```

```python
model.params.drop(['Intercept', 'trend']).plot(kind='bar', rot = 0)
plt.ylabel('Coefficient')
plt.show()
```

```python
model = sm.OLS.from_formula('y ~ trend', data = y.to_frame(name = 'y').assign(trend = np.arange(len(y)))).fit()
```

```python
def tsplot(y, lags = None, figsize = (10,8)):
    fig = plt.figure(figsize=figsize)
    layout = 2,2
    ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
    acf_ax = plt.subplot2grid(layout, (1, 0))
    pacf_ax = plt.subplot2grid(layout, (1, 1))
    y.plot(ax = ts_ax)
    smt.graphics.plot_acf(y, lags=lags, ax = acf_ax)
    smt.graphics.plot_pacf(y, lags=lags, ax = pacf_ax)
    [ax.set_xlim(1.5) for ax in [acf_ax, pacf_ax]]
    sns.despine()
    plt.tight_layout()
    return ts_ax, acf_ax, pacf_ax
```

```python
tsplot(model.resid, lags=4)
```

```python
y.to_frame(name = 'y').assign(dy = lambda x : x.y.diff()).plot(subplots=True)
sns.despine()
```

```python
data = (y.to_frame(name = 'y')
        .assign(dy = lambda df : df.y.diff())
        .assign(Ldy = lambda df : df.dy.shift())
       )
model = smf.ols(formula='dy ~ Ldy', data = data.dropna()).fit()
```

```python
tsplot(model.resid, lags = 4)
```

```python
smt.seasonal_decompose(y).plot()
```

$\mathrm{ARIMA}(p, d, q)$
$$y_t = c + \phi_1 y_{t-1} + \phi_2 y_{t-2} + \ldots + \phi_p y_{t-p} + e_t$$
$\Delta y_t = y_t - y_{t-1}$ for $d=1$

$$y_t = c + e_t + \theta_1 e_{t-1} + \theta_2 e_{t-2} + \ldots + \theta_q e_{t-q}$$

```python
model = smt.SARIMAX(y, trend = 'c', order = (1,1,1)).fit()
```

```python
tsplot(model.resid[2:], lags=4)
```

```python
model.summary()
```

$\mathrm{ARIMA}(p,d,q)(P,D,Q)_s$
$$\phi_p(L)\tilde{\phi}_P(L^S)\Delta^d\Delta_s^D y_t = A(t) + \theta_q(L)\tilde{\theta}_Q(L^s)e_t$$

```python
model = smt.SARIMAX(y, trend = 'c', order=(1, 1, 2), seasonal_order=(0, 1, 2, 12)).fit()
```

```python
model.summary()
```

```python
pred = model.get_prediction(start='2017-01-03')
pred_ci = pred.conf_int()
```

```python
ax = y.plot(label = 'Observed')
pred.predicted_mean.plot(ax = ax, label='Forecast', alpha = 0.7)
ax.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color='k', alpha =0.2)
```

```python
pred = model.get_prediction(start='2017-01-03', dynamic='2017-01-19')
pred_ci = pred.conf_int()
```

```python
ax = y.plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color='k', alpha=0.25)
ax.set_ylabel('Monthly Flight')

ax.fill_between(ax.get_ylim(), pd.Timestamp('2017-01-01'), y.index[-1], alpha=0.1, zorder=-1)
ax.annotate('Dynamic $\\longrightarrow$', (pd.Timestamp('2017-01-19'), 2030))
plt.legend()
sns.despine()
```
